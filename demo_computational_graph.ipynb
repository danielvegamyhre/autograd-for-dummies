```python
from autograd.engine import Scalar
from autograd.visualize import draw_graph
```

```python
# Simple computational graph
a = Scalar(2.5)
b = Scalar(-5.0)
c = (a**4) / 3
d = (2 * c) + (b**2)
d = d.relu()
d.backward()
draw_graph(d)
```

```python
# More complex computational graph (copied from https://github.com/karpathy/micrograd)

a = Scalar(-4.0)
b = Scalar(2.0)
c = a + b
d = a * b + b**3
c += c + 1
c += 1 + c + (-a)
d += d * 2 + (b + a).relu()
d += 3 * d + (b - a).relu()
e = c - d
f = e**2
g = f / 2.0
g += 10.0 / f

g.backward()

draw_graph(g)
```
